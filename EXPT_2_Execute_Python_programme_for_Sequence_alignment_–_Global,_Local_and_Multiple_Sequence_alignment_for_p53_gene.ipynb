```python
from Bio import Entrez
from Bio import SeqIO

Entrez.email = "dakshareddy26@gmail.com"
accession_number = "OF135708.1"

try:
    # Fetch the record from NCBI
    handle = Entrez.efetch(db="nucleotide", rettype="fasta", retmode="text", id=accession_number)
    fasta_sequence = handle.read()
    print(fasta_sequence)
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from Bio.Blast import NCBIWWW
from Bio import SeqIO

fasta_sequence = """
>OF135708.1 KR 1020210050308-A/6: Compositon comprising Tetragonia tetragonoides as an active ingredient for improving and treating Alzheimers disease
GTCGTGTACTCTCGGTAATACC
"""

result_handle = NCBIWWW.qblast("blastn", "nt", fasta_sequence)

with open("web_blast_output.xml", "w") as out_handle:
    out_handle.write(result_handle.read())

result_handle.close()

print("Web BLAST search completed. Results saved to web_blast_output.xml")
```

```python
from Bio.Blast import NCBIXML


blast_output_file = "web_blast_output.xml"

try:

    with open(blast_output_file, "r") as result_handle:

        blast_records = NCBIXML.parse(result_handle)


        for blast_record in blast_records:
            print(f"Query: {blast_record.query}")
            print(f"Database: {blast_record.database}")
            print(f"Number of hits: {len(blast_record.alignments)}")

            for alignment in blast_record.alignments:
                print(f"\n****Alignment****")
                print(f"Sequence: {alignment.title}")
                print(f"Length: {alignment.length}")
                print(f"E-value: {alignment.hsps[0].expect}")


except FileNotFoundError:
    print(f"Error: BLAST output file '{blast_output_file}' not found.")
except Exception as e:
    print(f"An error occurred during parsing: {e}")
```

```python
from Bio.Blast import NCBIXML
from Bio import SeqIO
import io

# The fetched FASTA sequence from the previous step
fasta_sequence = """
>OF135708.1 KR 1020210050308-A/6: Compositon comprising Tetragonia tetragonoides as an active ingredient for improving and treating Alzheimers disease
GTCGTGTACTCTCGGTAATACC
"""

# The BLAST output file from the web BLAST search
blast_output_file = "web_blast_output.xml"

print("--- Original FASTA Sequence ---")
print(fasta_sequence)
print("\n--- BLAST Results Summary ---")

try:
    # Open the BLAST output file
    with open(blast_output_file, "r") as result_handle:
        # Parse the XML output
        blast_records = NCBIXML.parse(result_handle)

        # Iterate through the BLAST records (usually there's only one for a single query)
        for blast_record in blast_records:
            print(f"Query: {blast_record.query}")
            print(f"Database: {blast_record.database}")
            print(f"Number of hits: {len(blast_record.alignments)}")

            # Iterate through the alignments (hits) and print key information
            for alignment in blast_record.alignments:
                print(f"\n  Sequence: {alignment.title}")
                print(f"  Length: {alignment.length}")
                # Accessing the first HSP (High-scoring Segment Pair)
                if alignment.hsps:
                    print(f"  E-value: {alignment.hsps[0].expect}")
                else:
                    print("  No HSPs found for this alignment.")


except FileNotFoundError:
    print(f"Error: BLAST output file '{blast_output_file}' not found.")
except Exception as e:
    print(f"An error occurred during parsing: {e}")
```
